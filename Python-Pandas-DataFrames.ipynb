```python
import pandas as pd

# Questão 1
dados_alunos = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana', 'Lucas'],
    'Idade': [15, 14, 16, 15, 14]
}

df_alunos = pd.DataFrame(dados_alunos)
display(df_alunos)
```

```python
import pandas as pd

# Questão 2
lista_rh = [
    {"Nome": "Ana", "Cargo": "Analista"},
    {"Nome": "Bruno", "Cargo": "Gerente", "Bonus": 500}
]

df_rh = pd.DataFrame(lista_rh)

# A coluna 'Bonus' para a funcionária Ana será preenchida com NaN (Not a Number),
# pois o dicionário original da Ana não possuía essa chave.
display(df_rh)
```

```python
import pandas as pd

# Questão 3
indices = ['p1', 'p2', 'p3']
nomes = pd.Series(['Notebook', 'Mouse', 'Teclado'], index=indices)
precos = pd.Series([3500, 150, 250], index=indices)

df_produtos = pd.DataFrame({'Produto': nomes, 'Preço': precos})
display(df_produtos)
```

```python
import pandas as pd

# Questão 4
data_vendas = {'Sede A': [1000, 2000, 1500], 'Sede B': [1200, 1800, 1100]}
df_vendas = pd.DataFrame(data_vendas)

# Criando novas colunas
df_vendas['Total Vendas'] = df_vendas['Sede A'] + df_vendas['Sede B']
df_vendas['Imposto'] = df_vendas['Sede A'] * 0.10

display(df_vendas)
```

```python
import pandas as pd

# Questão 5
df_clientes = pd.DataFrame({
    'Nome': ['Carlos', 'Beatriz'],
    'CPF': ['123.456.789-00', '987.654.321-11'],
    'Telefone': ['9999-8888', '7777-6666']
})

# Excluir CPF com del
del df_clientes['CPF']

# Excluir Telefone com pop() e armazenar
coluna_removida = df_clientes.pop('Telefone')

print("Conteúdo da coluna removida pelo pop():")
print(coluna_removida)
print("\nDataFrame final:")
display(df_clientes)
```

```python
import pandas as pd

# Questão 6
dados_paises = {
    'Capital': ['Brasília', 'Buenos Aires', 'Santiago'],
    'População': [214, 45, 19]
}
df_paises = pd.DataFrame(dados_paises, index=['Brasil', 'Argentina', 'Chile'])

# Recuperar dados do Brasil usando loc
print("Dados referentes ao Brasil:")
print(df_paises.loc['Brasil'])
```

```python
import pandas as pd

# Questão 7
carros = ['Fusca', 'Civic', 'Corolla', 'Onix', 'HB20', 'Sandero']
df_carros = pd.DataFrame({'Modelo': carros})

# Selecionar 4º (índice 3) e 5º (índice 4) carro usando iloc
print("Quarto e quinto carro:")
display(df_carros.iloc[3:5])
```

```python
import pandas as pd
import numpy as np

# Questão 8
# Criando DataFrame com 20 linhas
df_historico = pd.DataFrame({'Valores': np.random.randint(0, 100, 20)})

# Slicing da 5ª (índice 4) até a 12ª linha (índice 12 exclusive)
df_extraido = df_historico[4:12]
display(df_extraido)
```

```python
import pandas as pd

# Questão 9
df_loja1 = pd.DataFrame({'Produto': ['Arroz', 'Feijão'], 'Estoque': [50, 30]})
df_loja2 = pd.DataFrame({'Produto': ['Macarrão', 'Óleo'], 'Estoque': [20, 15]})

# Concatenar dados
df_consolidado = pd.concat([df_loja1, df_loja2], ignore_index=True)
display(df_consolidado)
```

```python
import pandas as pd

# Questão 10
df_func = pd.DataFrame({
    'Funcionário': ['Ana', 'Roberto', 'Clara', 'David'],
    'Salário': [5000, 4500, 7000, 3200]
})

print("1. Transposta:")
display(df_func.T)

print("\n2. Tipos de dados (dtypes):")
print(df_func.dtypes)

print(f"\n3. Quantidade total de elementos (size): {df_func.size}")

print("\n4. Duas últimas linhas (tail):")
display(df_func.tail(2))
```
